## (Predication) Model Building:

```python
X= data.iloc[:, 0:4].values
y = data.iloc[:, -1].values
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size= 0.2, random_state=12)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
model= LinearRegression()
print(model)
```

```python
train_model =model.fit(X_train,y_train)
print("model Training is completed")
print(train_model)
```

```python
## predication with Test data
y_pred =train_model.predict(X_test)
y_pred
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
mse = mean_squared_error(y_test,y_pred)
mae= mean_absolute_error(y_test,y_pred)
rmse =np.sqrt(mse)
print('mean squared error :',mse)
print('mean absolute_error:',mae)
print(rmse)
```

### R square 
### Formula
#### R^2 =1-SSR/SST
R^2 = Coefficient of determination SSR = Sum of square of residuals SST = total sum of squares.

```python
score=r2_score(y_test,y_pred)
print(score)
```

### Adjusted R2 =1 [(1-R2)*(n-1)/n-k-1]
where :

R2: The R2 of the model n: The number of observation k: The number of predictor variables.

```python
## display  adjusted R-squared
print(1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1))
```
